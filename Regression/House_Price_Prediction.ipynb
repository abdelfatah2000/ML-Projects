```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
boston_data = sklearn.datasets.load_boston()
print(boston_data)
```

Load Dataset to Pandas Dataframe

```python
data = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
```

```python
data.head()
```

Add target (price) column

```python
data['Price'] = boston_data.target
```

```python
data.head()
```

```python
data.shape
```

Check if there is a missing data


```python
data.isnull().sum()
```

```python
data.describe()
```

#Data Correlation

```python
correlation = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot=True, cbar=True, annot_kws={'size':9}, cmap="Blues", fmt=".2f")
```

```python
df_x = data.drop(["Price"], axis = 1)
df_y = data['Price']
```

Seperating data into Training data and Testing data 

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.18, random_state=42)
```

#Model Training

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
df = pd.DataFrame({'Actual' :y_test, 'Predicted': y_predict})
df.head(15)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test,y_predict)
print(MSE)
```

Visualizing the actual and predicted value

```python
plt.scatter(y_test,y_predict)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.show()
```
